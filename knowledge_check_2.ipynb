```python
# Import needed libraries

import pandas as pd
import numpy as mp

# Read in data from csv file

df = pd.read_csv('kpop_music_videos.csv')
df.head()
```

```python
# Replace NaN values in Director with "Unknown"

df['Director'].fillna('Unknown' , inplace = True)

df.head()
```

```python
# Replace NaN values in Korean Name with "No Korean Name"

df['Korean Name'].fillna('No Korean Name' , inplace= True)

df.head()
```

```python
# Replace string values in Type column with more suitable values

df['Type'] = df['Type'].replace(['Boy', 'Girl'], ['Boy Group', 'Girl Group'])

df.head()
```
